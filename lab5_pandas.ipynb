```python
data = {'name':['ram','gopal', 'verma'],
        'age':[25, 30, 45],
        'city':['bengaluru','delhi','mumbai']}
df = pd.DataFrame(data)

print(df["name"],"\n")
print(df.loc[0],"\n")
print(df.iloc[1],"\n")
print(df.at[1, "city"],"\n")
print(df[1:3],"\n")
df
```
